```python
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
from skimage import measure
```

```python
img = plt.imread("../data/cardiomegaly/45_IM-2081-1001.tiff")
mask = plt.imread("../data/cardiomegaly/45_IM-2081-1001_mask.png")
```

```python
contours = measure.find_contours(mask.mean(axis=2), .8)
```

```python
fig = plt.figure(figsize=(9, 9))
ax = plt.subplot(1, 1, 1)
ax.set_ylim(img.shape[0], 0)
ax.set_xlim(0, img.shape[1])
plt.imshow(img)

for contour in contours:
    plt.plot(contour[:, 1], contour[:, 0])
```
